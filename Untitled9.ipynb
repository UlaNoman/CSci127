```python
// Lab 6: LED PWM + Photoresistor logging
// Output CSV: time_ms, adc, pwm

const int LED_PIN = 9;     // PWM pin
const int LDR_PIN = A0;    // photoresistor divider to A0

// PWM settings
const int PWM_MIN = 0;
const int PWM_MAX = 255;

// Timing
const unsigned long SAMPLE_DT_MS = 6;   // sample every ~6 ms (like your data)
unsigned long lastSample = 0;

// Sine wave settings
const float FREQ_HZ = 1.0;             // 1 Hz oscillation (adjust if needed)
unsigned long t0 = 0;

void setup() {
  Serial.begin(115200);
  pinMode(LED_PIN, OUTPUT);
  t0 = millis();

  // Optional header (comment out if your grader doesnâ€™t want it)
  // Serial.println("t_ms,adc,pwm");
}

void loop() {
  unsigned long now = millis();
  if (now - lastSample < SAMPLE_DT_MS) return;
  lastSample = now;

  // time in seconds since start
  float t = (now - t0) / 1000.0;

  // sine between 0..1
  float s = 0.5 + 0.5 * sinf(2.0 * PI * FREQ_HZ * t);

  // map to PWM range
  int pwm = (int)(PWM_MIN + s * (PWM_MAX - PWM_MIN));
  pwm = constrain(pwm, 0, 255);

  analogWrite(LED_PIN, pwm);

  int adc = analogRead(LDR_PIN);

  // print CSV line: time_ms, adc, pwm
  Serial.print(now - t0);
  Serial.print(",");
  Serial.print(adc);
  Serial.print(",");
  Serial.println(pwm);
}
```
