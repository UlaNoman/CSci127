```python
// Spring photogate period measurement
const int TEMP_PIN  = A0;
const int LIGHT_PIN = A1;

const int THRESHOLD   = 830;
const int RESET_LEVEL = 845;
const unsigned long MIN_PERIOD_MS = 600;

bool armed = true;
unsigned long lastTrigger = 0;

void setup() {
  Serial.begin(9600);
  Serial.println("t_ms,A0_tempRaw,A1_lightRaw,period_s");
}

void loop() {
  int a0 = analogRead(TEMP_PIN);
  int a1 = analogRead(LIGHT_PIN);
  unsigned long now = millis();

  if (!armed && a1 > RESET_LEVEL) {
    armed = true;
  }

  if (armed && a1 < THRESHOLD) {
    armed = false;

    if (lastTrigger != 0) {
      unsigned long dt = now - lastTrigger;
      if (dt >= MIN_PERIOD_MS) {
        Serial.print(now); Serial.print(",");
        Serial.print(a0); Serial.print(",");
        Serial.print(a1); Serial.print(",");
        Serial.println(dt / 1000.0, 4);
        lastTrigger = now;
      }
    } else {
      lastTrigger = now;
    }
  }

  delay(2);
}
```
