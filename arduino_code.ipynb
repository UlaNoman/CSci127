```python
const int outputPin = 9;
const int inputPin  = A0;

void setup() {
  Serial.begin(9600);
  pinMode(outputPin, OUTPUT);
  tone(outputPin, 500);
}

void loop() {
  int voltage = analogRead(inputPin);
  Serial.println(voltage);
  delay(5);
}
```
